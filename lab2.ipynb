```python
import numpy as np

linear_data = np.arange(10, 99, 9) 

random_data = np.random.random(7)

print("Лінійний масив:", linear_data)
print("Випадковий масив:", random_data)

linear_data[0] = 1
random_data[0] = 0.99

print("Після оновлення:")
print("Лінійний масив:", linear_data)
print("Випадковий масив:", random_data)

linear_data = np.delete(linear_data, 5)
random_data = np.delete(random_data, 5)

print("Після видалення 5-го елемента:")
print("Лінійний масив:", linear_data)
print("Випадковий масив:", random_data)
```

```python
import pandas as pd
import numpy as np

data_numpy = np.random.rand(5, 3)  
df_numpy = pd.DataFrame(data_numpy, columns=['A', 'B', 'C'])

data_dict = {
    'Name': ['Anna', 'Bogdan', 'Roman', 'David', 'Eva'],
    'Age': [18, 22, 15, 40, 29],
    'Money': [50000, 60000, 70000, 80000, 65000]
}
df_dict = pd.DataFrame(data_dict)

print("DataFrame з NumPy масиву:")
print(df_numpy)

print("\nDataFrame зі словника:")
print(df_dict)

print("\nПерші рядки DataFrame з NumPy:")
print(df_numpy.head())

print("\nПерші рядки DataFrame зі словника:")
print(df_dict.head())

print("\nОписова статистика для DataFrame з NumPy:")
print(df_numpy.describe())

print("\nОписова статистика для DataFrame зі словника (тільки для числових даних):")
print(df_dict.describe())

print("\nОписова статистика для всіх даних DataFrame зі словника:")
print(df_dict.describe(include='all'))

print("\nВибір другого рядка з DataFrame зі словника за допомогою iloc:")
print(df_dict.iloc[1])


print("\nВибір рядка з ім'ям 'Roman' за допомогою loc:")
print(df_dict.loc[df_dict['Name'] == 'Roman'])

```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error

np.random.seed(42) 
x = np.linspace(0, 10, 100) 
true_y = 5 * x + 6

errors = np.random.normal(0, 1, size=x.shape)
measured_y = true_y + errors


mae = mean_absolute_error(true_y, measured_y)
mse = mean_squared_error(true_y, measured_y)

plt.figure(figsize=(8, 6))
plt.plot(x, true_y, label='', color='green', linestyle='--')
plt.scatter(x, measured_y, label='', color='blue', s=10)
plt.xlabel('x')
plt.ylabel('y')
plt.title(f'\nMAE = {mae:.2f}, MSE = {mse:.2f}')
plt.legend()
plt.grid(True)
plt.show()

df = pd.DataFrame({
    'x': x,
    'true_y': true_y,
    'measured_y': measured_y,
    'error': errors
})

df['MAE'] = mae
df['MSE'] = mse

csv_file_path = 'linear_function_with_errors.csv'
df.to_csv(csv_file_path, index=False)

print(f'Результати збережені у файл: {csv_file_path}')
```
